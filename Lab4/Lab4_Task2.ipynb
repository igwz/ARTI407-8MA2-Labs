# Lab 4 - Task 2
Apply histogram equalization to moon image and display the image with histogram.


```python
import matplotlib.pyplot as plt
from skimage import data, exposure, img_as_ubyte
```

```python
# Load moon image
img = data.moon()

# Histogram equalization (output is float in [0, 1])
img_eq = exposure.equalize_hist(img)
img_eq_u8 = img_as_ubyte(img_eq)

# Display images
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original moon image')
axes[0].axis('off')

axes[1].imshow(img_eq, cmap='gray')
axes[1].set_title('Equalized image')
axes[1].axis('off')
plt.tight_layout()
plt.show()

# Display histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(img.ravel(), bins=256, range=(0, 255), color='steelblue')
axes[0].set_title('Histogram - Original')
axes[0].set_xlabel('Gray level')
axes[0].set_ylabel('Pixel count')

axes[1].hist(img_eq_u8.ravel(), bins=256, range=(0, 255), color='seagreen')
axes[1].set_title('Histogram - Equalized')
axes[1].set_xlabel('Gray level')
axes[1].set_ylabel('Pixel count')
plt.tight_layout()
plt.show()
```
