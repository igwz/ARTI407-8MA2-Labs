# Lab 4 - Task 1
Rescale moon image intensities to the 3rd and 80th percentiles, then plot histograms.


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure
```

```python
# Load moon image
img = data.moon()

# Contrast stretching using 3rd and 80th percentiles
p3, p80 = np.percentile(img, (3, 80))
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

# Display images
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original moon image')
axes[0].axis('off')

axes[1].imshow(img_rescale, cmap='gray')
axes[1].set_title('Rescaled (3rd-80th percentile)')
axes[1].axis('off')
plt.tight_layout()
plt.show()

# Display histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(img.ravel(), bins=256, range=(0, 255), color='steelblue')
axes[0].set_title('Histogram - Original')
axes[0].set_xlabel('Gray level')
axes[0].set_ylabel('Pixel count')

axes[1].hist(img_rescale.ravel(), bins=256, range=(0, 255), color='darkorange')
axes[1].set_title('Histogram - Rescaled')
axes[1].set_xlabel('Gray level')
axes[1].set_ylabel('Pixel count')
plt.tight_layout()
plt.show()
```
