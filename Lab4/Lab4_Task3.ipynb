Use rocket image as reference and chelsea image as source, then perform histogram matching.


```python
import matplotlib.pyplot as plt
from skimage import data, exposure
```

```python
# Load source and reference images
source = data.chelsea()   # cat image
reference = data.rocket() # reference image

# Histogram matching
matched = exposure.match_histograms(source, reference, channel_axis=-1)

# Display source, reference, and matched images
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
axes[0].imshow(source)
axes[0].set_title('Source: chelsea')
axes[0].axis('off')

axes[1].imshow(reference)
axes[1].set_title('Reference: rocket')
axes[1].axis('off')

axes[2].imshow(matched)
axes[2].set_title('Matched result')
axes[2].axis('off')
plt.tight_layout()
plt.show()
```
