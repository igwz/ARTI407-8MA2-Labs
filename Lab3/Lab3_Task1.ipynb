```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Load image

```python
image_path = "Sunflower.jpg"

img = cv2.imread(image_path,0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

Increasing Image Size

```python
scale_factor = 2

scaled_img = cv2.resize(img, None, 
                        fx=scale_factor, 
                        fy=scale_factor, 
                        interpolation=cv2.INTER_CUBIC)

plt.imshow(scaled_img)
plt.title("Increased Image")
plt.axis("off")
plt.show()
```

Rotate Image by 120

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)

rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))

plt.imshow(rotated_img)
plt.title("Rotated Image (120)")
plt.axis("off")
plt.show()
```

X shear

```python
shear_factor = 0.5

M_shear_x = np.float32([
    [1, shear_factor, 0],
    [0, 1, 0]
])

sheared_x = cv2.warpAffine(img, M_shear_x, (int(w + shear_factor*h), h))

plt.imshow(sheared_x)
plt.title("X-shear")
plt.axis("off")
plt.show()
```

Y shear

```python
M_shear_y = np.float32([
    [1, 0, 0],
    [shear_factor, 1, 0]
])

sheared_y = cv2.warpAffine(img, M_shear_y, (w, int(h + shear_factor*w)))

plt.imshow(sheared_y)
plt.title("Y-shear")
plt.axis("off")
plt.show()
```
