# Lab 5 - Task 2
Apply 5x5 Gaussian and 21x21 Gaussian filters and show all images side by side.


```python
import cv2
import matplotlib.pyplot as plt
from skimage import data
```

```python
# Read image (image of choice)
image = data.coffee()  # RGB uint8 image

# Apply Gaussian filters with two kernel sizes
gaussian_5 = cv2.GaussianBlur(image, (5, 5), sigmaX=0)
gaussian_21 = cv2.GaussianBlur(image, (21, 21), sigmaX=0)

# Plot original + filtered images
fig, ax = plt.subplots(1, 3, figsize=(16, 5))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(gaussian_5)
ax[1].set_title('Gaussian Blur (5x5)')
ax[1].axis('off')

ax[2].imshow(gaussian_21)
ax[2].set_title('Gaussian Blur (21x21)')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
