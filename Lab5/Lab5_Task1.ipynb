# Lab 5 - Task 1
Convolve an image with a 7x7 box filter, then show original and smoothed images side by side.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
# Read image (image of choice)
image = data.astronaut()  # RGB uint8 image

# Create 7x7 box filter kernel
kernel = np.ones((7, 7), np.float32) / 49.0

# Convolve image with box filter
smoothed = cv2.filter2D(image, ddepth=-1, kernel=kernel)

# Plot original and smoothed images
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(smoothed)
ax[1].set_title('Smoothed Image (7x7 Box Filter)')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```
